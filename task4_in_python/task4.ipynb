```python
#1.What is the total number of candy ratings given across the three years. (Number of candy ratings, not the number of raters. Don’t count missing values)

df_q1 = df_tidy.dropna(axis = 0, subset = ['rate'])
df_q1['rate'].count()

#736095
```

```python
#2.What was the average age of people who are going out trick or treating?
df_q2 = df_tidy[df_tidy.going_out == 'Yes']
np.mean(df_q2.age)

#35.12
```

```python
#3.For each of joy, despair and meh, which candy bar received the most of these ratings?
```

```python
df_q1[df_q1.rate == 'DESPAIR'].groupby(['sweet']).count().sort_values('rate', ascending = False)
#Mary Janes has highest despair 
```

```python
df_q1[df_q1.rate == 'MEH'].groupby(['sweet']).count().sort_values('rate', ascending = False)
#Regular M&M has highest MEH rating 
```

```python
df_q1[df_q1.rate == 'JOY'].groupby(['sweet']).count().sort_values('rate', ascending = False)
#Regular M&M has highest JOY rating 
```

```python
#4.How many people rated Starburst as despair?
df_q4 = df_q1[df_q1.rate == 'DESPAIR']
df_q4[df_q4.sweet == 'Starburst'].count()
# 1864
```

```python
#5.What was the most popular candy bar by this rating system for each gender in the dataset ?
df_q5=df_q1.dropna(axis = 0, subset = ['sweet'])

rates=["JOY", "MEH", "DESPAIR"]
dummy_vars=[1,0, -1]

df_q5['point'] = df_q5['rate'].replace(rates, dummy_vars)
df_q5.head()

for g in df_q5['gender'].unique():
    print(g)
    print(df_q5[df_q5.gender == g ].groupby(['sweet']).sum().sort_values('point', ascending = False).head(1))
#it's probably not the best way of doing it 
#male/female/i'd rather not say all have favourite regular m&m.
```

```python
#6.What was the most popular candy bar in each year?
for y in df_q5['year'].unique():
    print(y)
    print(df_q5[df_q5.year == y ].groupby(['sweet']).sum().sort_values('point', ascending = False).head(1))
    
#2015 any full-size candy bar with point 4392
#2016 regular m&m with 1616
#2017 regular m&m with 4052
```

```python
#7.What was the most popular candy bar by this rating for people in US, Canada, UK, and all other countries?
country = ['UK', 'USA', 'CA']
for c in country:
    print(c)
    print(df_q5[df_q5.country == c ].groupby(['sweet']).sum().sort_values('point', ascending = False).head(1))
#all regular m&m 
```
